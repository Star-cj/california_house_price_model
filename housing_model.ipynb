```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# reading the data in data frame type 
df=pd.read_csv("housing.csv")
df
```

```python
# getting base information of the data frame
df.info()
```

```python
train_data=df
train_data
```

```python
train_data.hist(figsize=(10,7))
```

```python

train_data=train_data.drop(['ocean_proximity'],axis=1)
```

```python
from matplotlib.colors import ListedColormap

# creating my own colors 
colors=['orange','darkgreen','pink','blue','purple']
cmap=ListedColormap(colors)
plt.figure(figsize=(14,8))
sns.heatmap(train_data.corr(),annot=True, cmap=cmap)
# train_data.corr()
```

```python
# To turn the ocean_proximity to numerical value. 
numerical=pd.get_dummies(df['ocean_proximity'])
```

```python
# to include the ['ocean_proximity'] in the data and find the corrolation
train_data=train_data.join(numerical)
train_data
```

```python
# checking the color change in the corrolation
plt.figure(figsize=(15,8))
sns.heatmap(train_data.corr(),annot=True, cmap=cmap)
```

```python
# checking the corrolation among the lat. and long.
plt.figure(figsize=(9,6))
sns.scatterplot(x='latitude',y='longitude',data=train_data,hue="median_house_value",palette=cmap)
```

The purple represent the most expensive houses, and from there. You will see that they are very close to water

```python
train_data['bedroom_ratio']=train_data['total_bedrooms']/ train_data['total_rooms']
train_data['household_rooms']=train_data['total_rooms']/ train_data['households']
```

```python
# checking the corrolation with the new column added
plt.figure(figsize=(15,8))
sns.heatmap(train_data.corr(),annot=True, cmap=cmap)
```

```python
# to check if there is null value because model can't be train with missing or null value
nul=train_data.isna().sum()
nul
```

```python
# since the null value are in the columns that are int, so i replace it with zero 
train_data.fillna(0,inplace=True)
```

```python
# setting out 1/4 of the data for testing
from sklearn.model_selection import train_test_split
x=train_data.drop(['median_house_value'],axis=1)
y=train_data['median_house_value']
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.25)
```

```python
# we have to include the added features into our training data
from sklearn.linear_model import LinearRegression

reg=LinearRegression()
reg.fit(x_train,y_train)
```

```python
reg.score(x_test,y_test)
```

CHECKING FOR ANYWAY TO IMPROVE ACCURACY OF MODEL

```python
from sklearn.ensemble import RandomForestRegressor
ranForest=RandomForestRegressor()
ranForest.fit(x_train,y_train)
```

```python
ranForest.score(x_test,y_test)
```

```python
from sklearn.model_selection import GridSearchCV
# just know that GridSearchCV has made words in the dictionary to improve the model
grid= {
    "n_estimators": [100,200,300],
    "min_samples_split": [2,4],
    "max_features": [2,4,8]
}

gridSearch= GridSearchCV(ranForest,grid,cv=5,
                         scoring="neg_mean_squared_error",
                         return_train_score=True)

gridSearch.fit(x_train,y_train)
```

```python
bestEst=gridSearch.best_estimator_
bestEst
```
